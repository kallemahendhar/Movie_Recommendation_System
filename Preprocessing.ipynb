```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
# merging two data sets movies & credits
movies = movies.merge(credits,on ='title')
```

```python
# knowing the shape of the merged data set
movies.shape
```

```python
movies.head(1)
```

```python
movies.info()
```

```python
#required fields in the movies dataset:
#genres
#id
#keywords
#title
#overview
#cast
#crew

movies = movies[['id','title','overview','genres','keywords','cast','crew']]
```

```python

movies.head(1)
```

```python
# knowing how many null values are there in the dataset
movies.isnull().sum()
```

```python
#droping the null values
movies.dropna(inplace = True)
```

```python
# Knowing the duplicate values
movies.duplicated().sum()
```

```python
# converting the unformate list of dictionary to single list
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
# applying the covert def to genres
movies['genres'] = movies['genres'].apply(convert)
```

```python
#applying the covert def to keywords
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
#picking the top 3 values in the filed
import ast

def pickTop3Values(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count < 3:  # This ensures it collects up to 3 items
            L.append(i['name'])
            count += 1
        else:
            break
    return L
```

```python
#applying the pickTop3Values to cast filed
movies['cast'] = movies['cast'].apply(pickTop3Values)
```

```python
# taking director job in the crew
import ast

def director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
#applying director function
movies['crew'] = movies['crew'].apply(director)
```

```python
movies.head(1)
```

```python
# converting the overview string to list datatype
movies['overview'] = movies['overview'].apply(lambda x: x if isinstance(x, list) else x.split())
```

```python
movies.head(1)
```

```python
#removing the spaces in between words 
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head(1)
```

```python
#creating tags
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
#creating new dataset with the fileds of id,title,tags
new_df = movies[['id','title','tags']]
```

```python
new_df
```

```python
#coverting tags list datatype to string datatype
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
#coverting the tags into lower case
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
#applying stemming to covert every vector word into base word
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return ' '.join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#creating vectors
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
#Movie recommendation system model:
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0] #taking movie index which is user is given that movie we assume as usermovie
    distance = similarity[movie_index] #Taking all similarity distance of usermovie to all other movies including usermovie
    movies_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6] #sorted that distance in decending order and taking the top 5 movies except 1st one(because first one is the same movie) 
    for i in movies_list:
        print(new_df.iloc[i[0]].title) #printing all top 5 recommended movies title
```

```python
recommend('Avatar')
```
